# LOGISTIC REGRESSION

```python
def sigmoid(x):
    res = 1/(1+np.exp(-x))
    return res
```

```python
def Logistic_Regression(X, Y):
    
    k=.003 #learning rate
    m = X.shape[1] #No. of training samples
    n = X.shape[0] #No. of features
    W = np.zeros((n,1)) #Initializing weights
    B = 0               #Initializing bias
    cost_val = []     #for keeping track of the cost values at each iteration
    
    for i in range(30000):
        y_pred = sigmoid(np.dot(W.T, X) + B)
        cost = -(1/m)*np.sum( Y*np.log(y_pred) + (1-Y)*np.log(1-y_pred))       # cost function
        cost_val.append(cost)
        print("cost after ", i, "iteration is : ", cost)
        
        # Gradient Descent
        d_wrt_W = (1/m)*np.dot(y_pred-Y, X.T)
        d_wrt_B = (1/m)*np.sum(y_pred - Y)
        W = W - k*d_wrt_W.T
        B = B - k*d_wrt_B
        
    return W, B, cost_val
```

```python
def prediction(X,W,B):
    y_p = np.dot(W.T, X) + B
    y_pred = sigmoid(y_p)
    for i in range(y_pred.shape[1]):
        if (y_pred[0,i]<0.5):
            y_pred[0,i]=0
        else:
            y_pred[0,i]=1
    return y_pred
```

```python
def accuracy(X, Y, W, B):
    y_pred = prediction(X,W,B)
    acc = (1 - np.sum(np.absolute(y_pred - Y))/Y.shape[1])*100
    print("Accuracy of the model is : ",acc, "%")
```

```python
def prior_probability(data, Y):
    classes = sorted(list(data[Y].unique()))
    prior_probabilities = []
    for i in classes:
        m=len(data[data[Y]==i])
        n=len(data)
        prior_probabilities.append(m/n)
    return prior_probabilities
```

Fitting Normal Distribution

```python
def likelihood_calculation(data, attribute, attribute_val, Y, prob):
    data = data[data[Y]==prob]
    mean = data[attribute].mean()
    std = data[attribute].std()
    likelihood = (1 / (np.sqrt(2 * np.pi) * std)) *  np.exp(-((attribute_val-mean)**2 / (2 * std**2 ))) 
    return likelihood
```

```python
def naive_bayes(data, X, Y):
    attributes = list(data.columns)[:-1]
    prior = prior_probability(data, Y)
    Y_pred = []
    for x in X:
        classes = sorted(list(data[Y].unique()))
        likelihood = [1]*len(classes)
        for j in range(len(classes)):
            for i in range(len(attributes)):
                likelihood[j] = likelihood[j]*likelihood_calculation(data, attributes[i], x[i], Y, classes[j])

        # calculation of posterior probability
        post_prob = [1]*len(classes)
        for j in range(len(classes)):
            post_prob[j] = likelihood[j] * prior[j]
        Y_pred.append(np.argmax(post_prob))
    Y_pred=np.array(Y_pred) 
    return Y_pred
```

```python
Y_pred = naive_bayes(train, X=X_test, Y=34)
```
